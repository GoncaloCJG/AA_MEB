```python
# import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.pipeline import make_pipeline
```

```python
# Load datasets
neither_vs_melkeratosis_features_targets = pd.read_csv('../neither_vs_melkeratosis_features_targets.csv')
mel_vs_keratosis_features_targets = pd.read_csv('../mel_vs_keratosis_features_targets.csv')
fifty_neither_vs_mel_keratosis_features_targets = pd.read_csv('../fifty_neither_vs_mel_keratosis_features_targets.csv')
```

```python
neither_vs_melkeratosis_features_targets.head()
```

```python
# Train and test several models

# separate features and target
X = neither_vs_melkeratosis_features_targets.iloc[:,1:-2]
y = neither_vs_melkeratosis_features_targets['neither']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# make pipeline with scaling and model
pipe = make_pipeline(StandardScaler(), RandomForestClassifier())
pipe.fit(X_train, y_train)  # apply scaling on training data
pipe.score(X_test, y_test)  # apply scaling on testing data, without leaking training data.
```
