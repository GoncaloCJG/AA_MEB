```python
import os
import numpy as np
import shutil
```

```python
# images paths
melanoma_path = "../dataset_images/segmented/segmented/melanoma/"
neither_path = "../dataset_images/segmented/segmented/neither/"
keratosis_path = "../dataset_images/segmented/segmented/seborrheic keratosis/"

# folders to save images
melanoma_outs = "../dataset_images/segmented/melanoma_outs/"
neither_outs = "../dataset_images/segmented/neither_outs/"
keratosis_outs = "../dataset_images/segmented/keratosis_outs/"

if not os.path.exists(melanoma_outs):
    os.makedirs(melanoma_outs)
    
if not os.path.exists(neither_outs):
    os.makedirs(neither_outs)
    
if not os.path.exists(keratosis_outs):
    os.makedirs(keratosis_outs)

##create train, test and val for each class
#for melanoma
if not os.path.exists(melanoma_outs + "train/"):
    os.makedirs(melanoma_outs + "train/")
if not os.path.exists(melanoma_outs + "test/"):
    os.makedirs(melanoma_outs + "test/")
if not os.path.exists(melanoma_outs + "val/"):
    os.makedirs(melanoma_outs + "val/")
    
#for neither
if not os.path.exists(neither_outs + "train/"):
    os.makedirs(neither_outs + "train/")
if not os.path.exists(neither_outs + "test/"):
    os.makedirs(neither_outs + "test/")
if not os.path.exists(neither_outs + "val/"):
    os.makedirs(neither_outs + "val/")
    
#for keratosis
if not os.path.exists(keratosis_outs + "train/"):
    os.makedirs(keratosis_outs + "train/")
if not os.path.exists(keratosis_outs + "test/"):
    os.makedirs(keratosis_outs + "test/")
if not os.path.exists(keratosis_outs + "val/"):
    os.makedirs(keratosis_outs + "val/")
```

```python
# train, val, test split and save images in folders

## for melanoma
val_ratio = 0.10
test_ratio = 0.10

classe_directory = melanoma_path

allFileNames = os.listdir(classe_directory)
np.random.shuffle(allFileNames)
train_FileNames, val_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames) * (1 - (val_ratio + test_ratio))),
                                                           int(len(allFileNames) * (1 - val_ratio)),
                                                           ])

train_FileNames = [melanoma_path  + name for name in train_FileNames.tolist()]
val_FileNames = [melanoma_path  + name for name in val_FileNames.tolist()]
test_FileNames = [melanoma_path  + name for name in test_FileNames.tolist()]
    
# Copy-pasting images
for name in train_FileNames:
    shutil.copy(name, melanoma_outs + "train/")

for name in val_FileNames:
    shutil.copy(name, melanoma_outs + "val/")

for name in test_FileNames:
    shutil.copy(name, melanoma_outs + "test/")


    
## for neither
val_ratio = 0.10
test_ratio = 0.10

classe_directory = neither_path

allFileNames = os.listdir(classe_directory)
np.random.shuffle(allFileNames)
train_FileNames, val_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames) * (1 - (val_ratio + test_ratio))),
                                                           int(len(allFileNames) * (1 - val_ratio)),
                                                           ])

train_FileNames = [neither_path  + name for name in train_FileNames.tolist()]
val_FileNames = [neither_path  + name for name in val_FileNames.tolist()]
test_FileNames = [neither_path  + name for name in test_FileNames.tolist()]
    
# Copy-pasting images
for name in train_FileNames:
    shutil.copy(name, neither_outs + "train/")

for name in val_FileNames:
    shutil.copy(name, neither_outs + "val/")

for name in test_FileNames:
    shutil.copy(name, neither_outs + "test/")
    
    
    
## for keratosis
val_ratio = 0.10
test_ratio = 0.10

classe_directory = keratosis_path

allFileNames = os.listdir(classe_directory)
np.random.shuffle(allFileNames)
train_FileNames, val_FileNames, test_FileNames = np.split(np.array(allFileNames),
                                                          [int(len(allFileNames) * (1 - (val_ratio + test_ratio))),
                                                           int(len(allFileNames) * (1 - val_ratio)),
                                                           ])

train_FileNames = [keratosis_path  + name for name in train_FileNames.tolist()]
val_FileNames = [keratosis_path  + name for name in val_FileNames.tolist()]
test_FileNames = [keratosis_path  + name for name in test_FileNames.tolist()]
    
# Copy-pasting images
for name in train_FileNames:
    shutil.copy(name, keratosis_outs + "train/")

for name in val_FileNames:
    shutil.copy(name, keratosis_outs + "val/")

for name in test_FileNames:
    shutil.copy(name, keratosis_outs + "test/")
```

```python
# preparation of the dataset 

dataset_path = "../dataset_images/segmented/v1_multiclasse_dataset/"


if not os.path.exists(dataset_path):
    os.makedirs(dataset_path)

if not os.path.exists(dataset_path + "train/melanoma"):
    os.makedirs(dataset_path + "train/melanoma")
if not os.path.exists(dataset_path + "train/keratosis"):
    os.makedirs(dataset_path + "train/keratosis")
if not os.path.exists(dataset_path + "train/neither"):
    os.makedirs(dataset_path + "train/neither")
    

if not os.path.exists(dataset_path + "val/melanoma"):
    os.makedirs(dataset_path + "val/melanoma")
if not os.path.exists(dataset_path + "val/keratosis"):
    os.makedirs(dataset_path + "val/keratosis")
if not os.path.exists(dataset_path + "val/neither"):
    os.makedirs(dataset_path + "val/neither")
    
if not os.path.exists(dataset_path + "test/melanoma"):
    os.makedirs(dataset_path + "test/melanoma")
if not os.path.exists(dataset_path + "test/keratosis"):
    os.makedirs(dataset_path + "test/keratosis")
if not os.path.exists(dataset_path + "test/neither"):
    os.makedirs(dataset_path + "test/neither")")
    
train_img_paths = ["../dataset_images/segmented/keratosis_outs/train/",
                   "../dataset_images/segmented/melanoma_outs/train/",
                   "../dataset_images/segmented/neither_outs/train/"] 

val_img_paths = ["../dataset_images/segmented/keratosis_outs/val/",
                 "../dataset_images/segmented/melanoma_outs/val/",
                 "../dataset_images/segmented/neither_outs/val/"]

test_img_paths = ["../dataset_images/segmented/keratosis_outs/test/",
                  "../dataset_images/segmented/melanoma_outs/test/",
                  "../dataset_images/segmented/neither_outs/test/"]

for path in train_img_paths:
    imgs = os.listdir(path)
    
    for name in imgs:
        shutil.copy(path + name, dataset_path + "train/")
        
for path in val_img_paths:
    imgs = os.listdir(path)
    
    for name in imgs:
        shutil.copy(path + name, dataset_path + "val/")
        
for path in test_img_paths:
    imgs = os.listdir(path)
    
    for name in imgs:
        shutil.copy(path + name, dataset_path + "test/")

```
