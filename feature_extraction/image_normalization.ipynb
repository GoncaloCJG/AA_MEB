```python
# import modules
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
# Load an image and show it
img_true = cv.imread('../ISIC_0000001.jpg',-1)
img_true = cv.cvtColor(img_true, cv.COLOR_BGR2RGB) # Color conversion
plt.imshow(img_true); 
```

```python
# Show shape of the input image and create an array of the same shape
print(img_true.shape)

output_array = np.zeros((img_true.shape[0],img_true.shape[1]))
```

```python
# create a normalized image
img_norm = cv.normalize(img_true,  output_array, 0, 255, cv.NORM_MINMAX)
```

```python
# show normalized image
plt.imshow(img_norm)
```
